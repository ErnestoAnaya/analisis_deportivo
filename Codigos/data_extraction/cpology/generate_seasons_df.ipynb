### Obtener años


```python
def generate_seasons(first_year, last_year):
    '''
    Generar las temporadas especificadas que se quieren para una competencia y las guarda en el diccionario
    '''
    seasons = []
    
    for i in range(first_year, last_year+1):
        year_begin = i
        year_end = i+1
        
        season = str(year_begin)+'-'+str(year_end)
        
        seasons.append(season)
        
    return seasons
        
    
```

### Añadir competencias

```python
def add_comp(dict_,comp, seasons, teams):
    '''
    dict dict_: diccionario donde se van guardando los datos
    str comp: Nombre de la competencia
    str[]: lista de las temporadas de esa compatencia
    teams[][]: arreglos de equipos que compitieron en cada año
    
    Dada las competencias, inicializa el diccionario con toda la informacion dada
    '''
    #checar si no existe la competencia
    dict_comps = dict_.keys()
    if comp not in dict_comps:
        #verificar que coinciden las listas de temporadas y de equipos
        if len(seasons) == len(teams):
            content = {}
            
            for i in range(len(seasons)):
                content[seasons[i]] = teams[i]
        
            dict_[comp] = content
    else:
        "Competition already exists. Remove if you want to modify it"
```
