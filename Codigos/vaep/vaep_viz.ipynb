```python
import matplotsoccer
import pandas as pd
```

# VAEP vizualisations

```python
df = pd.read_csv('./xgboost_vaep.csv')
```

### Matplotsoccer test

https://github.com/TomDecroos/matplotsoccer

```python
actions = df[105:110]
```

```python
actions['short_name']
```

```python
matplotsoccer.actions(
    location=actions[["start_x", "start_y", "end_x", "end_y"]],
    action_type=actions.type_name,
    team=actions.team_name,
    result= actions.result_name == "success",
    label=actions[["time_seconds", "type_name", "short_name", "team_name", "vaep_value"]],
    labeltitle=["time","actiontype","player","team", "VAEP"],
    zoom=False
)
```
